```python
def deco(fn):
  def wrapper():
    print("before")
    fn()
    print("after")
  return wrapper

@deco
def hello():
  print("hello")

hello()

#Prediction: 1 line of code will print
#Prediction was incorrect. Printed 3 lines. Probably because I don't know how decorators work
```
